<h3>1. Generar código para obtener los términos del Corpus </h3>
Se programa utilizando la codificación "utf-8" y con insensibilidad a mayúsculas.

```python
#Se utiliza regex para eliminar signos de puntación y cáracteres especiales
import re

def limpiar_txt (txt):
    patron = r"[.,;\n\"]" 
    return re.sub(patron, '', txt)
```

```python
# Se construye el diccionario (términos del corpus)
path = "01_corpus_turismo.txt"

diccionario = []

corpus = open(path, "r", encoding="utf-8")
 
doc = corpus.readline()

while doc:
    for palabra in limpiar_txt(doc).split():
        palabra = palabra.lower()

        if not palabra in diccionario:
            diccionario.append(palabra)

    doc=corpus.readline()

corpus.close()

print (diccionario)
print (f"Cantidad de terminos: {len(diccionario)}")
```

```python
#Ordenar el diccionario de manera alfabética
diccionario_ordenado_az = diccionario.copy()
diccionario_ordenado_az.sort()
print("orden alfabético az:", diccionario_ordenado_az)


#Ordenar el diccionaro considerando la longitud de los términos
diccionaro_ordenado_long_terminos = diccionario.copy()

diccionaro_ordenado_long_terminos.sort(key=len)
print("orden alfabético long de términos:", diccionaro_ordenado_long_terminos)
```

<h3>2. Desarrollo de Matriz TF-IDF</h3>

```python
# Generar matriz término documento 
# Para ello, las filas representan los documentos y las columnas representa los términos (ordenados de manera álfabetico con sort())

corpus = open(path, "r", encoding="utf-8")
matriz_termino_documento = []

for doc in corpus:
    fila = [0 for x in range(len(diccionario_ordenado_az))] #se inicializa la fila para todas las columnas
  
    for j in range(len(diccionario_ordenado_az)):

        for palabra in limpiar_txt(doc).split():

            palabra = palabra.lower()

            if diccionario_ordenado_az[j] == palabra:
                fila[j]+= 1
    
    matriz_termino_documento.append(fila)
    
corpus.close()

print (matriz_termino_documento)
```

```python
# words   | doc1 |  doc2 | ....
#  2000   |      |       |
#  ....

def presentar_matriz(matriz, diccionario, index_id, round_val=1, len_col=6):
    aux = diccionario.copy()
    aux.sort(key=len)
    len_max_to_index = len(aux[-1]) # se ovtiene el tamaño más grande de las cadenas para formatear la tabla

    print(f"{index_id:^{len_max_to_index}}|", end="")

    for i in range(len(matriz)):
        val = "doc" + str(i+1)
        print(f"{val:^{len_col}}|", end="")


    for i in range(len(diccionario)):
        print(f"\n{diccionario[i]:^{len_max_to_index}}|", end="")

        for j in range(len(matriz)):
            print(f"{round(matriz[j][i],round_val ):^{len_col}}|", end="")
```

<p>IDF</p>

```python
# Cálculo de la Frecuencia Inversa De Documentos (IDF) -> Importancia global; mide cuán informativo es un término
import math 

len_max = len(diccionaro_ordenado_long_terminos[-1])
idf_t=[]
N = len(matriz_termino_documento)

for i in range(len(diccionario_ordenado_az)):
    #obtener df_t
    df_t = 0
    for j in range(N):
        if matriz_termino_documento[j][i] > 0:
            df_t +=1
    
    #calcular idf
    idf = math.log((N/df_t),10) 
    idf_t.append(idf)

print ("Valores IDF por termino")
for i in range(len(diccionario_ordenado_az)):
    print (f"{diccionario_ordenado_az[i]:^{len_max}}: {idf_t[i]}")
```

<p>Ponderación TF-IDF</p>

```python
#Ponderación TF-IDF
# w_t_d = tf_t_d * idf_t
w_t_d = []

for i in range(len(matriz_termino_documento)):

    wtd_i = [None for x in range(len(diccionario_ordenado_az))]

    for j in range(len(diccionario_ordenado_az)):
        val_tf_td = tf_t_d[i][j]
        val_idf_t = idf_t[j]
        val_wtd = val_tf_td * val_idf_t

        wtd_i[j]=val_wtd
    
    w_t_d.append(wtd_i)


print (w_t_d)
```
